```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from collections import Counter
plt.rcParams.update({'font.size': 22})
import sys
```

```python
#load catalog data in matchlocate format
path = '../../data/'
fl = 'dets_MAD20'
pt = '_nov18'
name = fl + pt
fmt = '.txt'
filename = path + name + fmt

data = pd.read_csv(filename, delim_whitespace=True, names=['id', 'date', 'time', 'lat', 'lon', 'dep', 'mag', 'cc', 'mad', 'ref'])

#delete unuseful columns in dataframe 

data.drop('cc', inplace=True, axis=1)
data.drop('mad', inplace=True, axis=1)
data.drop('ref', inplace=True, axis=1)
data.drop('id', inplace=True, axis=1)

#show new dataframe 

data.head()
```

```python
#kill negative magnitude detections

data = data[data['mag'] > 0]

data.head()
```

```python
#save dataframe to .csv format 

path = '../../data/' + name + '_arcmap.csv'
data.to_csv( path, index=True)
```
