```python
#We start loading and previewing our .txt data 

#set name will be added later, right now I'm working on a test file only

path = '../../data/'
file = 'dets_MAD20.txt'
fn = path + file
catalog = pd.read_csv(fn, delim_whitespace=True, names=['No', 'date', 'time', 'latitud', 'longitud', 'prof', 'mag', 'cc', 'mad', 'id'])

catalog.head()
```

```python
catalog[['dia', 'mes', 'anio']] = catalog['date'].str.split('/', expand=True)
catalog[['hora', 'min', 'sec']] = catalog['time'].str.split(':', expand=True)
catalog
```

```python
neworder = ['dia', 'mes', 'anio', 'hora', 'min', 'sec', 'latitud', 'longitud', 'prof', 'mag']
catalog = catalog.reindex(columns = neworder)
catalog
```
