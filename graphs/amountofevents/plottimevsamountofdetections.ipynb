```python
#turn pandas columns into lists to use in matplotlib

d = list(events['binedges'])

n = list(events['barheight'])


#get bar width from data 

w = (abs(events.diff(-1)))

wi = w.iloc[1,0]

print(wi)

```

```python
#setup and create bar plot

#Comment and uncomment as needed

fig, ax = plt.subplots(figsize=(15,15))

#plt.bar(d, n, align='edge', color='mediumslateblue', edgecolor='k')

plt.bar(d, n, align='edge', width=wi, color='mediumslateblue', edgecolor='k')


#for time of the day vs amount of detections

ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))

plt.xlabel('Hora (HH:MM:SS)')


#for date vs amount of detections

#plt.xlabel('Fecha (AAAA-MM)')

#for depth vs amount of detections

#plt.xlabel('Profundidad (km) ')


#for latitude vs amount of detections

#plt.xlabel('Latitud (°)')


#for longitude vs amount of detections

#plt.xlabel('Longitud (°)')


#for magnitude vs amount of detections
           
#plt.xlabel('Magnitud (Mw)')

           
           
plt.ylabel('Número de sismos')
plt.xticks(rotation=80)
plt.rcParams['xtick.labelsize'] = 18
```

```python
#calculate mean value of barheight to get 'normal' seismicity levels

mbh = statistics.mean(n)
sdbh = statistics.stdev(n)
upl = mbh + sdbh
lwl = mbh - sdbh


events['baselevel1'] = lwl
events['baselevel2'] = upl
print(mbh, sdbh)
events.head()
```

```python
#create bar plot marking 'normal' seismicity levels 

#Comment and uncomment as needed
fig, ax = plt.subplots(figsize=(15,15))

plt.bar(d, n, align='edge', width=wi, color='mediumslateblue', edgecolor='k')

plt.axhspan(lwl, upl, facecolor='crimson', alpha=0.2, label='nivel promedio de sismicidad')


#for time of the day vs amount of detections

ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))

plt.xlabel('Hora (HH:MM:SS)')


#for date vs amount of detections

#plt.xlabel('Fecha (AAAA-MM)')

#for depth vs amount of detections

#plt.xlabel('Profundidad (km)')


#for latitude vs amount of detections

#plt.xlabel('Latitud (°)')


#for longitude vs amount of detections

#plt.xlabel('Longitud (°)')


#for magnitude vs amount of detections
           
#plt.xlabel('Magnitud (Mw)')
           
           
plt.ylabel('Número de sismos')
plt.xticks(rotation=80)
plt.rcParams['xtick.labelsize'] = 18
plt.legend()
```
